```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
#fetching the data
url = 'https://sih-temper-app.eu-gb.mybluemix.net/getuserdb'   #put the required url
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys = True)
```

```python
data = JSONContent['table']
data
```

```python
df = pd.DataFrame(data)
df.columns = ['uid', 'city' , 'state', "age_range_id", "latitude", "longitude", "si_rank", "se_rank","cs_rank","rt_rank","overall_rank"]
df.fillna(0 , inplace =  True)
df.head()
```

```python
# def si_per():
#     value = (df.si_rank/7) * 100
#     return value
# df
df['si_rank'] = df['si_rank'].apply(lambda x: ((x/7) * 100) )
df['se_rank'] = df['se_rank'].apply(lambda y: ((y/4) * 100))
df['cs_rank'] = df['cs_rank'].apply(lambda z : ((z/6) * 100))
df['rt_rank'] = df['rt_rank'].apply(lambda a : ((a/3) * 100))
df['overall_rank'] = df['overall_rank'].apply(lambda b : ((b/20) * 100))
num = df._get_numeric_data()
num[num < 0] = 0
df.head()
```
