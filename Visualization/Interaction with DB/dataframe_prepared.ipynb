```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
#fetching the data
url = 'https://sih-temper-app.herokuapp.com/getuserdb'   #put the required url
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys = True)
```

```python
data = JSONContent['table']
data
```

```python
df = pd.DataFrame(data)
df.columns = ['uid', 'city' , 'state', "age_range_id", "latitude", "longitude", "se_rank", "rt_rank","cs_rank","si_rank","overall_rank"]
df.head()
```

```python
new_df = df.assign(se_per = (df.se_rank/4)*100, rt_per = (df.rt_rank/3)*100,
                   cs_per = (df.cs_rank/6)*100, si_per = (df.si_rank/7)*100, overall_per = (df.overall_rank/20)*100)
new_df
```
