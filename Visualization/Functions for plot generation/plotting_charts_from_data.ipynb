```python
#importing required libraries
import numpy as np
import pandas as pd
import requests
import json
```

```python
#fetching the data
url = 'https://sih-temper-app.herokuapp.com/getuserdb'   #put the required url
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys = True)
print(content)
```

```python
data = JSONContent['table']
data
```

```python
#creating dataframe
df = pd.DataFrame(data)
df
```

```python
df.columns = ['uid', 'city' , 'state', "age_range_id", "latitude", "longitude", "se_rank", "rt_rank","cs_rank","si_rank","overall_rank"]
df.head()
```
